<h1>Databricks Clusters</h1>

<h2>Create and Manage Interactive Clusters</h2>
<p>
A Databricks cluster is a set of computation resources and configurations on which you run data engineering, data science, and data analytics workloads, such as production ETL pipelines, streaming analytics, ad-hoc analytics, and machine learning. You run these workloads as a set of commands in a notebook or as an automated job.
</p>

<p>Databricks makes a distinction between all-purpose clusters and job clusters.</p>
<ul>
<li>You use all-purpose clusters to analyze data collaboratively using interactive notebooks.</li>

<li>You use job clusters to run fast and robust automated jobs.</li>
</ul>
<p>This demo will cover creating and managing all-purpose Databricks clusters using the Databricks Data Science & Engineering Workspace.</p>


<h3>Learning Objectives</h3>
<p>By the end of this lesson you'll be able to:</p>
<ul>
<li>Use the Cluster UI to configure and deploy a cluster</li>
<li>Edit, terminate, restart and delete clusters</li>
</ul>

<h3>Create Cluster</h3>

<p>Depending on the workspace in which you're currently working, you may or may not have cluster creation privileges.</p>

<p>Instructions in this section assume that you do have cluster creation privileges, and that you need to deploy a new cluster to execute the lessons in this course.</p>

<p><strong>NOTE:</strong> Check with your instructor or a platform admin to confirm whether or not you should create a new cluster or connect to a cluster that has already been deployed. Cluster policies may impact your options for cluster configuration.</p>

<h4>Steps:</h4>

<ol>
<li>Use the left sidebar to navigate to the Compute page by clicking on the compute icon</li>

<li>Click the blue Create Cluster button</li>

<li>For the Cluster name, use your name so that you can find it easily and the instructor can easily identify it if you have problems</li>

<li>Set the Cluster mode to Single Node (this mode is required to run this course)</li>

<li>Use the recommended Databricks runtime version for this course</li>

<li>Leave boxes checked for the default settings under the Autopilot Options</li>

<li>Click the blue Create Cluster button</li>
</ol>

<p><strong>NOTE:</strong> Clusters can take several minutes to deploy. Once you have finished deploying a cluster, feel free to continue to explore the cluster creation UI.</p>

<h3>Single-Node Cluster Required for This Course</h3>
<p><strong>IMPORTANT</strong>:This course requires you to run notebooks on a single-node cluster.</p>
<p>Follow the instructions above to create a cluster that has <strong>Cluster mode</strong> to <strong>Single Node</strong></p>
